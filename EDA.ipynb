### Function "col_to_bar" converts each column to a bar chart visualization of count per category

```python
def stat_dict (data_name, col_name):
    out = {}
    d = data_name[col_name]
    categories = d.unique()[1:]

    for c in categories:
        if c is np.nan:
            out['nan'] = sum(d.isna())
        else:
            out[str(c)] = len(d[d==c])

    return out
```

```python
def col_to_bar (data_name, col_name):

    d = stat_dict(data_name, col_name)
    d = dict(sorted(d.items(), key=lambda x:x[1]))
    
    keys = d.keys()
    values = d.values()

    plt.bar(keys, values)

    plt.title('Per Category Count')
    plt.xlabel('Categories')
    plt.ylabel('Quantity')
    plt.xticks(rotation=45)

    plt.show()
```
